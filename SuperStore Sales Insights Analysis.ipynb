Check the Missing Values

```python
# Check for missing values
missing_values = data.isnull().sum()
print("Missing values in each column:\n", missing_values)

# Drop rows with any missing values (if any)
data_cleaned = data.dropna()

# Check for duplicates
duplicates = data_cleaned.duplicated().sum()
print("Number of duplicate rows:", duplicates)

# Drop duplicates
data_cleaned = data_cleaned.drop_duplicates()

# Display the first few rows of the cleaned DataFrame
data_cleaned.head()
```

# Arrays

```python
import numpy as np

# Convert Sales, Profit, and Discount columns to a NumPy array
sales_profit_discount = data[['Sales', 'Profit', 'Discount']].values

# Verify the conversion
print(sales_profit_discount)
```

# Statastical Operations

```python
# Calculate the mean of Sales, Profit, and Discount
mean_values = np.mean(sales_profit_discount, axis=0)
print("Mean values:", mean_values)

# Calculate the median
median_values = np.median(sales_profit_discount, axis=0)
print("Median values:", median_values)

# Calculate the standard deviation
std_values = np.std(sales_profit_discount, axis=0)
print("Standard Deviation:", std_values)
```

#  Conditional Operations

```python
# Apply a condition to identify transactions with a discount greater than 20%
high_discount = sales_profit_discount[sales_profit_discount[:, 2] > 0.2]
print("High Discount Transactions:", high_discount)

# Set Profit to 0 for transactions with a Sales value below 100
sales_profit_discount[:, 1] = np.where(sales_profit_discount[:, 0] < 100, 0, sales_profit_discount[:, 1])
print("Updated Profit with condition:", sales_profit_discount[:, 1])
```

# Aggregation Operations

```python
# Sum of all Sales
total_sales = np.sum(sales_profit_discount[:, 0])
print("Total Sales:", total_sales)

# Maximum Profit
max_profit = np.max(sales_profit_discount[:, 1])
print("Maximum Profit:", max_profit)
```

# Random Sampling

```python
# Generate random indices to sample data
random_indices = np.random.choice(sales_profit_discount.shape[0], 5, replace=False)
random_sample = sales_profit_discount[random_indices]
print("Random Sample of Transactions:", random_sample)

# Shuffle the dataset
np.random.shuffle(sales_profit_discount)
print("Shuffled Data:", sales_profit_discount)
```

```python
# 2. Sales and Profit by Region
import pandas as pd
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
region_sales_profit = data.groupby('Region').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
region_sales_profit = pd.melt(region_sales_profit, id_vars=['Region'], value_vars=['Sales', 'Profit'])
sns.barplot(x='Region', y='value', hue='variable', data=region_sales_profit, palette='coolwarm')
plt.title('Sales and Profit by Region')
plt.ylabel('Amount')
plt.show()
```

Insights from the Grouped Bar Plot:
Sales vs. Profit by Region:

The grouped bar plot allows you to compare Sales and Profit side by side for each region.
Sales Bars: Higher bars for sales indicate regions where the total sales are greater.
Profit Bars: The profit bars show the total profit earned in each region. Comparing these with the sales bars gives insight into profitability.
